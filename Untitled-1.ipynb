# Project Proposal : Global Weather & Air Quality Analysis

## 1-Dataset Identification
* **Dataset Name:** Global Weather Repository
* **Source Link:**(https://www.kaggle.com/datasets/nelgiriyewithana/global-weather-repository)
* **Data Provider:** **Kaggle**

## 2- Dataset Description
* **Size:** The dataset consists of **121,212 records (rows)** and **41 features (columns)**.
* **General Idea:** is an all-encompassing daily overview of weather situation and air quality indices in the large cities of any country.
* **Feature Breakdown:**
    * **Spatial Data:** Country, location name, latitude, longitude, and timezone.
    * **Weather Metrics:** Temperature (C/F), weather condition text, atmospheric pressure, and humidity.
    * **Wind & Visibility:** Wind speed, direction, gust speeds, and visibility range.
    * **Air Quality Indicators:** Concentrations of $CO, O_3, NO_2, SO_2, PM_{2.5},$ and $PM_{10}$.
    * **Astronomy:** Sunrise/sunset times, moon phases, and illumination levels.
* **Real-world Significance:** This is a real-world dataset because it captures live, fluctuating environmental data that impacts public health, aviation, and agriculture globally.

## 3- Evidence of Raw/Unclean Data
This dataset is **not clean** and requires significant wrangling based on the following observations:

1.  **Redundant Data (Duplicates in Logic):** The data contains the same measurements in different units (e.g., `temperature_celsius` vs `temperature_fahrenheit` and `wind_mph` vs `wind_kph`), which creates multi-collinearity issues for AI models.
2.  **Inconsistent Categorical Labels:** The `condition_text` column has inconsistent casing (e.g., "Partly Cloudy" vs "Partly cloudy"), meaning the same weather state is treated as two different categories.
3.  **Strange Formatting (Object Types):** Temporal columns like `last_updated`, `sunrise`, and `sunset` are stored as **Strings (Object)** instead of **DateTime** objects, making time-series analysis impossible without transformation.
4.  **Mixed Formats:** The wind direction is categorical (NNW, SSE), while its degree is numerical, requiring synchronization.
5.  **Potential Outliers:** Air quality sensors often record extreme spikes (Outliers) during specific environmental events that need to be identified and handled.

## 4. Main Analytical Questions
The project aims to answer the following key questions:
1.  **Environmental Correlation:** What is the relationship between temperature, humidity, and air pollutants like Carbon Monoxide (CO)?
2.  **Global Pollution Hotspots:** Which countries exhibit the highest average $PM_{2.5}$ levels, and how does weather (like rain or wind) affect these levels?
3.  **Predictive Insight:** Can we categorize the "Air Quality Status" effectively based on meteorological variables?


## 5. Motivation - Why this Dataset?
* **Technical Challenge:** The dataset's size (121k+ rows) and the variety of its features (categorical, numerical, temporal) make it perfect for demonstrating advanced data cleaning and visualization skills.
* **Real-world Impact:** Improving air quality starts with understanding the data. This project simulates a real-time monitoring system that can provide actionable insights for environmental safety.
